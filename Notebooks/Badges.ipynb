```python
#Importing Libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
headers = ['rowId',"userId","name","date","class","tagBased"]
df = pd.read_csv('Badges.csv', header = 0)
df.head()
```

```python
df.columns = headers
```

```python
df.head()
```

```python
df['name'].value_counts()
```

```python
df['name'].dtype
```

```python
#Grouping Name elements based on Value Counts

name_grp = df[['name']].groupby(['name']).size()
```

```python
print(name_grp)
```

```python
name_grp.plot(kind='bar', figsize=(15,8), rot=90)
plt.xlabel("Name of Badge")
plt.ylabel("Frequency")
plt.title("Bar Plot of the Value Counts of Badges")
plt.show()
```

### Analysis 1: Highest Frequency Badges

- **Autobiographer** and **Yearling** have **highest frequency** followed by supporter.
- **Autobiographer** belongs to the **Participation** Group and can be obtained by completing the **"About Me"** section.
- **Yearling** belongs to the **Participation** Group and can be obtained by gaining **200 reputation**.
- **Supporter** belongs to the **Moderation** Group and can be obtained by providing **1 upvote**.

This happens because these 3 badges are the easiest to obtain as the user has to put lesser efforts in obtaining them. This is supported by the fact that **Autobiographer** and **Supporter** are **Bronze** Badge(easy to get) whereas **Yearling** is a **Silver** Badge(A little difficult to get.)

```python
name_grp.plot(kind='bar', figsize=(15,8), rot=90, log=True)
plt.xlabel("Name of Badge")
plt.ylabel("Frequency")
plt.title("Bar Plot (Log) of the Value Counts of Badges")
plt.show()
```

### Analysis 2: Some Badges lower than highest frequency

- The next level of Badges is for **Nice Answer, Nice Question, Student and Teacher**.
- **Nice Answer** and **Nice Question** have almost **similar frequency** can be understood from the fact that the method to obtain them is almost the same. For Nice Question/Answer user needs to get **10 points(1 upvote)** on his Question/Answer respectively.
- Similarly **Student** and **Teacher** badges can be obtained by **Asking a Question** and getting a **score of 1 or more** and **Answering a Question** with a **score of 1 or more** respectively.

An important point to note from the above analysis is that all **4 badges** belong to the **Bronze** Category.

### Analysis 3: Lowest Frequency Badges

- Least frequent badges are Copy Editor, Refiner, Research Assistant and Socratic.
